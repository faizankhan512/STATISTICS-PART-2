1.What is hypothesis testing in statistics?

 ANS>> Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves testing an assumption (the null hypothesis) against an alternative explanation (alternative hypothesis) using sample evidence.

2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

ANS>> Null Hypothesis (H₀): A default statement that there is no effect, no difference, or no relationship in the population.

Alternative Hypothesis (H₁ or Ha): It contradicts the null, suggesting there is an effect, a difference, or a relationship.

Example:
H₀: μ = 100 (The population mean is 100)
H₁: μ ≠ 100 (The population mean is not 100)

3.What is the significance level in hypothesis testing, and why is it important?

ANS>>Denoted by α (alpha), it represents the threshold probability for rejecting the null hypothesis.

Common values: 0.05, 0.01

It defines the risk of making a Type I error (rejecting a true H₀).

4.What does a P-value represent in hypothesis testing?

ANS>>The P-value is the probability of obtaining the observed results (or more extreme) assuming that the null hypothesis is true.

5.How do you interpret the P-value in hypothesis testing?

ANS>>P ≤ α: Reject the null hypothesis (significant result).

P > α: Fail to reject the null hypothesis (not significant).

6.What are Type I and Type II errors in hypothesis testing?

ANS>>Type I Error (α): Rejecting a true null hypothesis (false positive).

Type II Error (β): Failing to reject a false null hypothesis (false negative).

7.What is the difference between a one-tailed and a two-tailed test?

ANS>>One-tailed test: Tests for an effect in one direction (e.g., H₁: μ > 100).

Two-tailed test: Tests for an effect in both directions (e.g., H₁: μ ≠ 100).

8.What is the Z-test, and when is it used in hypothesis testing?

ANS>>The Z-test is used when:

The sample size is large (n > 30).

The population standard deviation is known.

It compares the sample mean to the population mean.

Between one sample mean and a known value (one-sample t-test).

Example: CI = (98.1, 101.9) means we are 95% confident the true mean is between 98.1 and 101.9.

```python
#1.Write a Python program to perform a Z-test for comparing a sample mean to a
# known population mean and interpret the results.
import numpy as np
from scipy import stats

sample = [105, 98, 100, 102, 101, 99, 97, 103, 100, 104]

population_mean = 100
population_std = 5

sample_mean = np.mean(sample)
sample_size = len(sample)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Sample Mean:", sample_mean)
print("Z-Score:", z_score)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

```python
5.# Create a Python function that calculates and visualizes Type 1 and Type 2
  # errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def show_type1_type2():
    # Settings
    mu0 = 100       # Null hypothesis mean
    mu1 = 105       # Alternative hypothesis mean
    sigma = 10      # Standard deviation
    n = 30          # Sample size
    alpha = 0.05    # Significance level

    se = sigma / np.sqrt(n)  # Standard error
    z_crit = norm.ppf(1 - alpha / 2)  # Critical z-value for 95% CI

    # Critical sample mean values
    lower = mu0 - z_crit * se
    upper = mu0 + z_crit * se

    # X values
    x = np.linspace(90, 115, 1000)
    h0 = norm(mu0, se).pdf(x)
    h1 = norm(mu1, se).pdf(x)

    # Plot
    plt.plot(x, h0, label='H0', color='blue')
    plt.plot(x, h1, label='H1', color='green')

    # Type I error (red)
    plt.fill_between(x, 0, h0, where=(x < lower) | (x > upper), color='red', alpha=0.3, label='Type I Error')

    # Type II error (purple)
    plt.fill_between(x, 0, h1, where=(x >= lower) & (x <= upper), color='purple', alpha=0.3, label='Type II Error')

    plt.axvline(lower, color='black', linestyle='--')
    plt.axvline(upper, color='black', linestyle='--')

    plt.title("Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function
show_type1_type2()
```

```python
# 6.Write a Python program to perform an independent T-test and interpret the
# results
import numpy as np
from scipy import stats

# Sample data for two independent groups
group1 = [85, 90, 88, 75, 95]
group2 = [80, 70, 78, 85, 76]

# Perform independent T-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Print results
print("T-statistic:", round(t_statistic, 3))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (means are significantly different).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference between means).")

```

```python
# 7.Perform a paired sample T-test using Python and visualize the comparison
# results.
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Paired sample data (before and after treatment, for example)
before = [85, 90, 88, 75, 95]
after = [88, 92, 89, 78, 97]

# Perform paired T-test
t_stat, p_value = stats.ttest_rel(before, after)

# Print results
print("T-statistic:", round(t_stat, 3))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant change).")
else:
    print("Result: Fail to reject the null hypothesis (no significant change).")

# Visualization
plt.plot(before, label="Before", marker='o')
plt.plot(after, label="After", marker='o')
plt.title("Paired Sample Comparison")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 8.Simulate data and perform both Z-test and T-test, then compare the results
# using Python.
import numpy as np
from scipy import stats

# Simulate random sample data
np.random.seed(1)
sample = np.random.normal(loc=102, scale=10, size=20)  # mean=102, std=10, n=20

# Known population values for Z-test
population_mean = 100
population_std = 10

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)
sample_std = np.std(sample, ddof=1)  # sample standard deviation

# Z-test
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
z_p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# T-test
t_stat, t_p_value = stats.ttest_1samp(sample, population_mean)

# Results
print("=== Z-test ===")
print("Z-Score:", round(z_score, 3))
print("P-Value:", round(z_p_value, 4))

print("\n=== T-test ===")
print("T-Statistic:", round(t_stat, 3))
print("P-Value:", round(t_p_value, 4))

# Comparison interpretation
alpha = 0.05
if z_p_value < alpha:
    print("\nZ-test: Significant difference (reject H0)")
else:
    print("\nZ-test: No significant difference (fail to reject H0)")

if t_p_value < alpha:
    print("T-test: Significant difference (reject H0)")
else:
    print("T-test: No significant difference (fail to reject H0)")
```

```python
# 12.Perform a Chi-square test for independence between two categorical
# variables in Python.
import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table
# Rows: Gender (Male, Female)
# Columns: Preference (Product A, Product B)
data = [[30, 10],   # Male
        [20, 40]]   # Female

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Print results
print("Chi-square Statistic:", round(chi2, 3))
print("Degrees of Freedom:", dof)
print("P-value:", round(p, 4))
print("\nExpected Frequencies Table:")
print(np.round(expected, 2))

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nResult: Variables are dependent (reject H0)")
else:
    print("\nResult: Variables are independent (fail to reject H0)")
```

```python
# 16.Implement an F-test using Python to compare the variances of two random
# samples.
import numpy as np
from scipy.stats import f

# Step 1: Simulate two random samples
np.random.seed(1)
sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=52, scale=8, size=30)

# Step 2: Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Step 3: Compute F-statistic (larger variance / smaller variance)
if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1
else:
    F = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1

# Step 4: Calculate p-value (two-tailed)
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

# Step 5: Print results
print("Sample 1 Variance:", round(var1, 2))
print("Sample 2 Variance:", round(var2, 2))
print("F-Statistic:", round(F, 3))
print("P-Value:", round(p_value, 4))

# Step 6: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Variances are significantly different (reject H0).")
else:
    print("Result: No significant difference in variances (fail to reject H0).")
```

```python
# 20.D Perform a two-way ANOVA test using Python to study the interaction
# between two factors and visualize the results.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Create a simple dataset
data = {
    'Score': [75, 78, 74, 70, 82, 79, 77, 73, 85, 88, 84, 90],
    'TeachingMethod': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F']
}
df = pd.DataFrame(data)

# Two-way ANOVA
model = ols('Score ~ C(TeachingMethod) + C(Gender) + C(TeachingMethod):C(Gender)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("Two-way ANOVA Result:\n")
print(anova_table)

# Visualization: Interaction plot
sns.pointplot(data=df, x='TeachingMethod', y='Score', hue='Gender', markers=['o', 's'], capsize=0.1)
plt.title("Interaction Plot: Teaching Method vs Score by Gender")
plt.grid(True)
plt.show()
```

```python
# 22. Perform a one-way ANOVA test in Python and visualize the results with
# boxplots to compare group means.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

# Step 1: Create example data for 3 groups
group1 = [85, 90, 88, 75, 95]
group2 = [80, 70, 78, 85, 76]
group3 = [92, 88, 91, 94, 90]

# Step 2: Perform one-way ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

# Step 3: Print results
print("F-statistic:", round(f_stat, 3))
print("P-value:", round(p_value, 4))

# Step 4: Interpretation
if p_value < 0.05:
    print("Result: Significant difference between group means (reject H0)")
else:
    print("Result: No significant difference between group means (fail to reject H0)")

# Step 5: Visualization with boxplot
# Combine data into a DataFrame
df = pd.DataFrame({
    'Score': group1 + group2 + group3,
    'Group': ['Group 1']*len(group1) + ['Group 2']*len(group2) + ['Group 3']*len(group3)
})

# Plot boxplot
sns.boxplot(x='Group', y='Score', data=df)
plt.title("Boxplot of Scores by Group")
plt.grid(True)
plt.show()
```

```python
# 24.Perform a hypothesis test for population variance using a Chi-square
# distribution and interpret the results.
import numpy as np
from scipy.stats import chi2

# Step 1: Sample data
sample = [10, 12, 9, 11, 13, 10, 12, 11, 14, 10]
n = len(sample)
sample_var = np.var(sample, ddof=1)

# Step 2: Hypothesized population variance (H0: sigma^2 = 4)
hypo_var = 4

# Step 3: Chi-square statistic
chi_square_stat = (n - 1) * sample_var / hypo_var

# Step 4: Critical values for 95% confidence (alpha = 0.05, two-tailed)
alpha = 0.05
chi2_lower = chi2.ppf(alpha / 2, df=n - 1)
chi2_upper = chi2.ppf(1 - alpha / 2, df=n - 1)

# Step 5: Interpretation
print("Sample Variance:", round(sample_var, 2))
print("Chi-square Statistic:", round(chi_square_stat, 3))
print("Chi-square Critical Range: [{:.3f}, {:.3f}]".format(chi2_lower, chi2_upper))

if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Result: Reject the null hypothesis (variance significantly different).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference in variance).")
```

```python
# 25.Write a Python script to perform a Z-test for comparing proportions between
# two datasets or groups.
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Example data
# Group 1: 45 successes out of 100
# Group 2: 30 successes out of 90
success_counts = np.array([45, 30])
sample_sizes = np.array([100, 90])

# Perform two-proportion Z-test
z_stat, p_value = proportions_ztest(success_counts, sample_sizes)

# Print results
print("Z-statistic:", round(z_stat, 3))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Significant difference in proportions (reject H0).")
else:
    print("Result: No significant difference in proportions (fail to reject H0).")
```

```python
# 26.Implement an F-test for comparing the variances of two datasets, then
# interpret and visualize the results.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Step 1: Create two datasets
np.random.seed(0)
group1 = np.random.normal(loc=50, scale=5, size=30)   # std = 5
group2 = np.random.normal(loc=52, scale=8, size=30)   # std = 8

# Step 2: Calculate sample variances
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# Step 3: Calculate F-statistic (larger variance / smaller variance)
if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(group1) - 1, len(group2) - 1
else:
    F = var2 / var1
    dfn, dfd = len(group2) - 1, len(group1) - 1

# Step 4: Calculate p-value (two-tailed)
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

# Step 5: Output
print("Variance of Group 1:", round(var1, 2))
print("Variance of Group 2:", round(var2, 2))
print("F-statistic:", round(F, 3))
print("P-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("Result: Significant difference in variances (reject H0)")
else:
    print("Result: No significant difference in variances (fail to reject H0)")

# Step 6: Visualize the F-distribution
x = np.linspace(0.1, 5, 500)
y = f.pdf(x, dfn, dfd)

plt.plot(x, y, label='F-distribution')
plt.axvline(F, color='red', linestyle='--', label=f'F-stat = {round(F, 2)}')
plt.title("F-Test for Comparing Two Variances")
plt.xlabel("F-value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

```
